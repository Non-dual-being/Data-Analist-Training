```python
age_groups = [(x,x+10) for x in (range(20,86,10))]

```

```python
age_groups = [(x+1, y) if x > 20 else (x, y) for (x, y) in age_groups]


age_groups
```

```python
def assing_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <=age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return('Unknown')
```

```python
food_2['Age_Groups'] = food_2['Age'].apply(assing_age_group)
```

```python


# Voeg een kolom toe die aangeeft of 'Cmp_Total_Accepted' groter is dan 0
food_2['Accepted_Status'] = food_2['Cmp_Total_Accepted'].apply(lambda x: 'Accepted' if x > 0 else 'Not Accepted')

```

```python


# Bereken de sommen voor elke 'Purchases' kolom, gegroepeerd op 'Accepted_Status'
Summed_2 = pd.DataFrame({
    'Purchases_Type': Purchases,
    'Sum_Accepted': food_2[food_2['Accepted_Status'] == 'Accepted'][Purchases].sum(),
    'Sum_Not_Accepted': food_2[food_2['Accepted_Status'] == 'Not Accepted'][Purchases].sum()
}).reset_index(drop=True)

# Bekijk het resultaat
Summed_2
```

```python
import matplotlib.pyplot as plt

# Maak een stacked bar chart
Summed_2.set_index('Purchases_Type')[['Sum_Accepted', 'Sum_Not_Accepted']].plot(kind='bar', stacked=True)

# Voeg labels en titel toe
plt.xlabel('Purchases Type')
plt.ylabel('Sum')
plt.title('Sum of Purchases by Acceptance Status')

# Toon de grafiek
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Stel de Seaborn stijl in
sns.set(style="whitegrid")

# Maak een lege plot voor de gestapelde bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Plot de gestapelde balken met matplotlib
ax.bar(Summed_2['Purchases_Type'], Summed_2['Sum_Accepted'], label='Accepted', color='b')
ax.bar(Summed_2['Purchases_Type'], Summed_2['Sum_Not_Accepted'], bottom=Summed_2['Sum_Accepted'], label='Not Accepted', color='r')

# Voeg labels en titel toe
ax.set_xlabel('Purchases Type')
ax.set_ylabel('Sum')
ax.set_title('Sum of Purchases by Acceptance Status')

# Voeg een legenda toe
ax.legend()

# Toon de grafiek
plt.xticks(rotation=45)
plt.show()
```
