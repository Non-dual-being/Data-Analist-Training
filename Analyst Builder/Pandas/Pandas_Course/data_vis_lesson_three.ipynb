```python
path = r"C:\Users\Stichting GeoFort\Downloads\smoking.csv"
smoking = pd.read_csv(path)
smoking
```

```python
smoking = smoking[smoking['smoking'] == 'Yes']
```

```python
female_smokers = smoking[smoking['gender'] == 'Female']
male_smokers = smoking[smoking['gender'] == 'Male']
```

```python
fig, axs = plt.subplots(1,2)

sns.histplot(female_smokers['age'], bins = 25, ax= axs[0])

axs[0].set_title('Age Distribution - Female Smokers')
axs[1].set_title('Age Distribution - Male Smokers')

sns.histplot(male_smokers['age'], bins = 25, ax= axs[1])
plt.tight_layout()
plt.show()
```

```python
df_corr = df.drop(columns='zip_code')
```

```python
correlation = df_corr.corr(numeric_only=True)
```

```python
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

```python
pearson = df_corr.corr(numeric_only=True, method='pearson')
```

```python
spearman, kendall = df_corr.corr(numeric_only=True, method='spearman'), df_corr.corr(numeric_only=True, method='kendall')
```

```python
kendall
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

sns.heatmap(pearson, annot=True, cmap='coolwarm', ax=axs[0], cbar=False)
sns.heatmap(spearman, annot=True, cmap='coolwarm', ax=axs[1], cbar=False)
sns.heatmap(kendall, annot=True, cmap='coolwarm', ax=axs[2])

# Zet de titels
axs[0].set_title('Pearson heatmap')
axs[1].set_title('Spearman heatmap')
axs[2].set_title('Kendall heatmap')

# Verwijder de y-as bij de eerste twee plots
axs[2].set_yticks([])  # Verwijdert de y-as ticks
axs[2].set_ylabel('')   # Verwijdert de y-as labels

axs[1].set_yticks([])  # Verwijdert de y-as ticks
axs[1].set_ylabel('')   # Verwijdert de y-as labels


axs[2].set_xticklabels(axs[2].get_xticklabels(), rotation=0)
sns.set_style('darkgrid')

plt.tight_layout()
plt.show()
```
