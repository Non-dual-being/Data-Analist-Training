```python
import pandas as pd
import os
```

```python
path_1 = os.getcwd()
relative_path = r"datasets\banking.csv"
path = os.path.join(path_1, relative_path)
```

```python
bank = pd.read_csv(path)
bank
```

```python
bank.groupby('education')['outstanding_balance'].min().reset_index()
```

```python
bank.groupby('education')['outstanding_balance'].max().reset_index()
```

```python
bank.groupby('education')['outstanding_balance'].sum().sort_values(ascending=False).reset_index()
```

```python
bank.groupby('education')['outstanding_balance'].count().sort_values(ascending=False).reset_index()
```

```python
bank.groupby('education')['outstanding_balance'].median().sort_values(ascending=False).reset_index()
```

```python
result = bank.groupby('education')['outstanding_balance'].agg(['count','sum','min','max']).reset_index()
```

```python
result['avg'] = result['sum']/result['count']
result
```

```python
bank.groupby('education').agg({'outstanding_balance':['sum','count','mean']}).reset_index().sort_values(by=('outstanding_balance', 'mean'), ascending=False)
```

```python
bank.groupby('education').agg(sum_balance = ('outstanding_balance','sum'), count_balance = ('outstanding_balance','count')).reset_index().sort_values(by='sum_balance', ascending=False)
```

```python
result = bank.groupby(['job','marital']).agg({'outstanding_balance':'sum'}).reindex()
```

```python
result.loc['management']
```

```python
result.loc[('management','single')]
```
