```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"C:\Users\Stichting GeoFort\Downloads\all_stocks_5yr.csv"

df = pd.read_csv(path)
```

```python
company_AAL = df[df['Name'] == 'AAL']
```

```python

plt.close('all')

# Vervolgens kun je opnieuw je plot aanroepen
company_AAL['close'].plot.hist(bins=10, figsize=(10, 8))
plt.title('Histogram AAPL Close Values')  # Gebruik de juiste functie plt.title()
plt.xlabel('Closing Values')  # plt.xlabel() voor de x-as
plt.ylabel('Frequency')  # plt.ylabel() voor de y-as
plt.show()
```

```python
company_AAL['close'].plot.hist(bins=10, figsize=(10, 8), label='Frequenty bars', color='green', linestyle='--')
plt.title('Histogram AAPL Close Values', fontdict={'fontname':'Times New Roman','fontsize':'20'})  # Gebruik de juiste functie plt.title()
plt.xlabel('Closing Values')  # plt.xlabel() voor de x-as
plt.ylabel('Frequency')  # plt.ylabel() voor de y-as

plt.yticks([x for x in range(0,301,50)])
plt.legend()
plt.savefig(path)
plt.show()

```
