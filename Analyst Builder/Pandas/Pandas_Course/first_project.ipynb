```python
df_2 = df.dropna(subset='genres')
```

```python
df_2.loc[:,'profit'] = df_2.loc[:,'revenue'] - df_2.loc[:,'budget']
```

```python
movies = df_2.drop(columns=['cast','homepage','director','tagline','keywords','overview'])
```

```python
movies.head()
```

```python
movies.drop(columns=['production_companies','release_year','budget_adj','revenue_adj'], inplace=True)
```

```python
movies.head()
```

```python
movies.drop(columns=['id','imdb_id'], inplace=True)
```

```python
movies.head()
```

```python
from pandas import Series, DataFrame
movies['genres'].str.split('|').apply(Series,1)#Series converts the list from the splut tot a serie, 1 indicates the axis as column
```

```python
split = movies['genres'].str.split('|').apply(Series,1).stack()
```

```python
split
```

```python
split.index = split.index.droplevel(-1)
```

```python
split
```

```python
split.name = 'genres_split'
split
```

```python
del movies['genres']
movies = movies.join(split)
```

```python
movies
```

```python
movies_genres = movies.groupby('genres_split')['original_title'].count().sort_values(ascending=False).reset_index(name='count')
```

```python
movies_genres.set_index('genres_split', inplace=True)
movies_genres
```

```python
# alternatief
genres_count = pd.DataFrame(movies.groupby('genres_split')['original_title'].nunique()).sort_values('original_title', ascending=False).rename(columns={'original_title':'number_of_movies'})
```

```python
genres_avg = pd.DataFrame(movies.groupby('genres_split').mean(numeric_only=True))
genres_avg
```

```python
pd.options.display.float_format = '{:2f}'.format
```

```python
genres_budget = genres_avg.sort_values('budget', ascending=True)
genres_revenue = genres_avg.sort_values('revenue', ascending=True)
genres_profit = genres_avg.sort_values('profit', ascending=True)
```

```python
genres_budget[['budget','revenue']].plot.barh(title='sharon', figsize=(10,10), color = ['darkblue', 'c'])
```

```python
genres_revenue[['budget','revenue']].plot.barh(title='sharon', figsize=(10,10), color = ['darkblue', 'c'])
```

```python
import matplotlib.pyplot as plt

# Maak een figuur en assen aan
fig, ax1 = plt.subplots(figsize=(10, 10))

# Plot de 'profit' op de eerste as (ax1)
genres_profit['profit'].plot.barh(ax=ax1, color='darkblue', position=0, width=0.4)
ax1.set_xlabel('Profit')
ax1.set_title('Sharon')

# Maak een tweede as (ax2) die dezelfde y-as deelt
ax2 = ax1.twiny()

# Plot de 'popularity' op de tweede as (ax2)
genres_profit['popularity'].plot.barh(ax=ax2, color='c', position=1, width=0.4)
ax2.set_xlabel('Popularity')
plt.tight_layout()
plt.show()
```

```python
movies_3 = df_2.drop(columns=['cast','homepage','director','tagline','keywords','overview','production_companies','budget_adj','revenue_adj'])
```

```python
movies_3.drop(columns=['id','imdb_id'],inplace=True)
```

```python
split_2= movies_3['genres'].str.split('|').apply(Series,1).stack()
split_2.index = split_2.index.droplevel(-1)
split_2.name = 'genres_split'
del movies_3['genres']
movies_3=movies_3.join(split_2)
```

```python
movies_3.head()
```

```python
sharon = pd.DataFrame(movies_3.groupby(['release_year','genres_split'])['profit'].mean())
```

```python
sharon_pivoted = pd.pivot_table(sharon, values='profit', index='genres_split', columns='release_year')
sharon_pivoted
```

```python
sns.heatmap(sharon_pivoted, cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(sharon_pivoted, cmap='YlGnBu')
plt.plot(figsize=(16,16))
plt.show()
```
