```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Hypothetische data gebaseerd op maand data
dagen = np.array(range(1, 31)).reshape(-1, 1)
prijzen = np.array([4.00, 4.10, 4.15, 4.12, 4.20, 4.18, 4.25, 4.30, 4.28, 4.35, 4.40, 4.38, 4.45, 4.50, 4.48, 4.55, 4.60, 4.58, 4.65, 4.70, 4.68, 4.75, 4.80, 4.78, 4.85, 4.90, 4.88, 4.95, 5.00, 4.50])

# Lineaire regressie model
model = LinearRegression()
model.fit(dagen, prijzen)

# Voorspellen van de volgende 30 dagen
toekomstige_dagen = np.array(range(31, 61)).reshape(-1, 1)
voorspelde_prijzen = model.predict(toekomstige_dagen)

# Visualisatie
plt.figure(figsize=(10, 5))
plt.scatter(dagen, prijzen, color='blue', label='Historische Prijzen')
plt.plot(toekomstige_dagen, voorspelde_prijzen, color='red', linestyle='--', label='Voorspelde Prijzen')
plt.xlabel('Dag')
plt.ylabel('Prijs (EUR)')
plt.title('Voorspelling van UNFI Prijzen')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Basic statistics
unfi_statistics = unfi.describe()
unfi_statistics
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Split data into training and test sets
train_data = unfi['close'][:int(0.8*len(unfi))]
test_data = unfi['close'][int(0.8*len(unfi)):]

# Fit the ARIMA model
model = ARIMA(train_data, order=(5, 1, 0))
model_fit = model.fit()

# Make predictions
forecast = model_fit.forecast(steps=len(test_data))

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data, label='Training Data')
plt.plot(test_data.index, test_data, label='Test Data', color='red')
plt.plot(test_data.index, forecast, label='Forecast', color='green', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.title('ARIMA Model Forecast for UNFI Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from textblob import TextBlob
import matplotlib.pyplot as plt

# Voorbeeld dataset van teksten
data = {
    'text': [
        "UNFI is gearing up for a bullish wave.",
        "UNFI dead project.",
        "I am very bullish on this stinky project.",
        "UNFI is showing strong growth potential.",
        "This is a scam, stay away from UNFI.",
        "UNFI is undervalued and a great investment opportunity.",
        "Negative news about UNFI is spreading.",
        "Positive developments are coming for UNFI.",
        "UNFI has solid fundamentals and a strong team.",
        "I have lost all hope in UNFI."
    ]
}

# Laad data in een DataFrame
df = pd.DataFrame(data)

# Functie om sentiment scores te berekenen met TextBlob
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Pas sentimentanalyse toe op alle teksten
df['sentiment'] = df['text'].apply(get_sentiment)

# Toon de resultaten
df

# Visualiseren van de resultaten
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['sentiment'], color=df['sentiment'].apply(lambda x: 'green' if x > 0 else 'red'))
plt.xlabel('Text Index')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Analysis of UNFI Related Texts')
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.show()
```

```python
import itertools
import numpy as np
from sklearn.metrics import mean_squared_error

# Parameter ranges
p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))

# Find the best ARIMA parameters
best_aic = np.inf
best_order = None
best_mdl = None

for order in pdq:
    try:
        model = ARIMA(train_data, order=order)
        model_fit = model.fit()
        aic = model_fit.aic
        if aic < best_aic:
            best_aic = aic
            best_order = order
            best_mdl = model_fit
    except:
        continue

print(f'Best ARIMA order: {best_order} with AIC: {best_aic}')

# Use the best model to forecast
forecast = best_mdl.forecast(steps=len(test_data))

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data, label='Training Data')
plt.plot(test_data.index, test_data, label='Test Data', color='red')
plt.plot(test_data.index, forecast, label='Forecast', color='green', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.title('ARIMA Model Forecast for UNFI Prices')
plt.legend()
plt.grid(True)
plt.show()
```
