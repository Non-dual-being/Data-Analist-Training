```python
import re
```

```python
customer['Phone_Number'].apply(
    lambda x: x[0:4] + x[4:7] + "-" + x[-3:] if isinstance(x, str) and not bool(re.match(r'(?i)^na[n]?-*$', x)) else x
)
```

-   `re.match():` Deze functie zoekt naar een match aan het begin van de string.
-   `r'(?i)^na[n]?-*$':` Deze regex matcht zowel 'nan' als 'na', ongeacht hoofdlettergevoeligheid, en gevolgd door nul of meer streepjes.
-   `not bool(re.match(...)):` Dit controleert of de regex niet matcht, wat betekent dat de string geen 'nan' of 'na' bevat.

```python
customer.loc[:, 'Phone_Number'] = customer['Phone_Number'].apply(
    lambda x: x[0:4] + x[4:7] + "-" + x[-3:] if isinstance(x, str) and not bool(re.match(r'(?i)^na[n]?-*$', x)) else x
)
```

- np.nan: Dit is de waarde die wordt teruggegeven als de conditie waar is.

- pd.isna(x) or pd.Series(x).str.contains(r'(?i)^na[n]?-*$', regex=True).any(): Dit is de conditie die wordt geëvalueerd. Het gebruikt or om te controleren of x een NaN is of voldoet aan de regex-patroon.

- x: Dit is de waarde die wordt teruggegeven als de conditie onwaar is.
